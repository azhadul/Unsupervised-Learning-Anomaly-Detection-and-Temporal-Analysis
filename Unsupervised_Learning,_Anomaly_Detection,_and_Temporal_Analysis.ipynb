Question 1 : What is Dimensionality Reduction? Why is it important in machine
learning?
-   Dimensionality reduction is the process of reducing the number of input features (variables) in a dataset while keeping as much important information as possible​
It is important in machine learning because it reduces model complexity, helps avoid overfitting, speeds up training and prediction, and makes high‑dimensional data easier to visualize and understand

Question 2: Name and briefly describe three common dimensionality reduction
techniques..
Common dimensionality reduction techniques include:

Principal Component Analysis (PCA): A linear technique that creates new uncorrelated features (principal components) which capture most of the variance in the data, and then keeps only the top few components.​

Linear Discriminant Analysis (LDA): A supervised method that finds projections which best separate classes while reducing dimensionality.​

t‑SNE (t‑Distributed Stochastic Neighbor Embedding): A nonlinear method mainly used for visualizing high‑dimensional data in 2D or 3D by preserving local neighborhood structure.

Question 3: What is clustering in unsupervised learning? Mention three popular
clustering algorithms.
-  Clustering is an unsupervised learning task where data points are grouped into clusters such that points in the same cluster are more similar to each other than to points in other clusters.​
Three popular clustering algorithms are:

   K‑means clustering: Partitions the data into
   k clusters by assigning points to the nearest centroid and iteratively updating centroids.​

   DBSCAN (Density‑Based Spatial Clustering of Applications with Noise): Forms clusters based on dense regions of points and labels isolated points as noise.​

   Hierarchical clustering: Builds a hierarchy of clusters (often visualized with a dendrogram) by successively merging or splitting clusters.

Question 4: Explain the concept of anomaly detection and its significance.
- Anomaly detection is the process of identifying data points, events, or observations that deviate significantly from the normal pattern in a dataset.​
It is significant because anomalies often correspond to important events such as fraud, equipment failures, cyber‑attacks, or data quality issues, so detecting them early helps prevent loss and improve reliability and safety.​



Question 5: List and briefly describe three types of anomaly detection techniques
-  Three main types of anomaly detection techniques are:

   Statistical methods: Assume a probabilistic model for normal data (for example, Gaussian) and flag points with very low probability or lying far from the mean as anomalies.​

   Distance or density‑based methods: Use distances or densities (for example, k‑nearest neighbors, DBSCAN, Local Outlier Factor) and mark points in sparse, low‑density regions as anomalies.​

   Machine‑learning–based methods: Use models such as one‑class SVM, isolation forest, or autoencoders trained on normal data to detect points that do not fit the learned normal pattern

Question 6: What is time series analysis? Mention two key components of time series
data.
-  Time series analysis is the study of data collected over time in order to understand its underlying patterns (such as trend and seasonality) and to make forecasts.​

   Two key components of time series data are:

   Trend: The long‑term direction of movement in the data, either upward, downward, or stable over time.​

   Seasonality: Regular, repeating patterns at fixed periods such as daily, weekly, monthly, or yearly effects

Question 7: Describe the difference between seasonality and cyclic behavior in time
series.

-  Seasonality refers to regular and predictable patterns that repeat at a fixed and known period, such as higher sales every December or higher power usage every summer.​

   Cyclic behavior refers to up‑and‑down movements that occur over longer, variable lengths of time and are not tied to a fixed calendar period, such as multi‑year business cycles.

Question 8: Write Python code to perform K-means clustering on a sample dataset.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 1. Create a sample 2D dataset
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.5, random_state=42)

# 2. Define KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# 3. Fit the model to the data
kmeans.fit(X)

# 4. Get predicted cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print("Cluster labels for first 10 points:", labels[:10])
print("Cluster centroids:\n", centroids)
```

Question 9: What is inheritance in OOP? Provide a simple example in Python.
-  Inheritance in Object‑Oriented Programming (OOP) is a mechanism where a new class (child/subclass) reuses and extends the properties and methods of an existing class (parent/superclass).​

   It promotes code reuse and makes it easier to create specialized classes that add or override behavior from a base class

```python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("The animal makes a sound")

# Child class inheriting from Animal
class Dog(Animal):
    def speak(self):
        print(self.name, "barks")

# Using the classes
a = Animal("Generic")
a.speak()        # The animal makes a sound

d = Dog("Tommy")
d.speak()        # Tommy barks
```

Question 10: How can time series analysis be used for anomaly detection?
-   Time series analysis can be used for anomaly detection by first modeling the normal temporal patterns (trend, seasonality, and typical variation) and then flagging observations that deviate strongly from model predictions.​
Examples include using forecasting models (such as ARIMA or exponential smoothing) or decomposition to obtain expected values over time and then marking points with unusually large residuals (errors) as anomalies.
